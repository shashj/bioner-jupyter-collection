````python
from vllm import LLM, SamplingParams
````

````python
llm = LLM(model="meta-llama/Meta-Llama-3-8B-Instruct", dtype='float16')
````

````python
# llm_quant = LLM(model="astronomer/Llama-3-8B-Instruct-GPTQ-8-Bit",
#           quantization='gptq', dtype='float16')
````

````python
tokenizer = llm.get_tokenizer()
````

````python
tokenizer.eos_token_id
````

````python
tokenizer.convert_tokens_to_ids("<|eot_id|>")
````

````python
prompts = [
    "Hello, my name is",
    "The president of the United States is",
    "The capital of France is",
    "The future of AI is",
]
sampling_params = SamplingParams(temperature=0.1, top_p=0.35, max_tokens=500, stop_token_ids=[tokenizer.eos_token_id, tokenizer.convert_tokens_to_ids("<|eot_id|>")])
````

````python
!pip install ipython-autotime
````

````python
%load_ext autotime
outputs = llm.generate(prompts, sampling_params)

# Print the outputs.
for output in outputs:
    prompt = output.prompt
    generated_text = output.outputs[0].text
    print(f"Prompt: {prompt!r}, Generated text: {generated_text!r}")
````

````python
prompt = """

You are a medical coder and you are extracting disease from below charts. Extract the diseases and give an output in the form of a python list.

Things to keep in mind while extracting diseases:

1. Extract them exactly as mentioned in the chart, do not modify
2. Extract the longest mention of the disease, do not cut short
3. Don't add any explanation or any other text
4. Make sure to give a python list, it will be read by a program

Patient chart is given below in backticks.

```

Suchitra has following diseases:

1. Hypertension unspecified
2. Type 2 diabetes mellitus with multiple complications

```

"""
````

````python
conversations = tokenizer.apply_chat_template(
    [{'role': 'user', 'content': prompt}],
    tokenize=False,
)
conversations
````

````python
%load_ext autotime

outputs = llm.generate([conversations], sampling_params)
````
